```python
TRAIN_DF_PATH = "./output/train_df.pickle"
PREDICT_DF_PATH = "./output/predict_df.pickle"
MERIDIAN_PATH = "../../data/life/Bite Off More Than You Can Chew Braces profile - Sheet.csv"
```

```python
FUTURE_BRACE_START_INDEX = 14
FUTURE_BRACE_END_INDEX = 43
```

```python
import os
import sys
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

utils_path = os.path.abspath('../../utils')
sys.path.insert(0, utils_path)

from teeth_utils import normalise_tooth_positions
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
train_df = pd.read_pickle(TRAIN_DF_PATH)
predict_df = pd.read_pickle(PREDICT_DF_PATH)
```

```python
# training

models = []
# each model belongs to a domain
for row in train_df.iterrows():
    x = [v[0].flatten() for v in row[1]]
    y = [v[1] for v in row[1]]
    model = LinearRegression()
    model.fit(x, y)
    models.append(model)
```

```python
# predict

ys = []
for i, row in predict_df.iterrows():
    x = [v.flatten() for v in row]
    y = models[i].predict(x)
    ys.append(y)
```

```python
column_labels = ["Brace " + str(i) for i in range(FUTURE_BRACE_START_INDEX, FUTURE_BRACE_END_INDEX)]
```

```python
ys = np.array(ys)
future_df = pd.DataFrame(data = ys, columns = column_labels)
future_df
```

```python
past_df = pd.read_csv(MERIDIAN_PATH)
past_df = past_df.dropna(subset=["Tooth number", "Domain"])

# wip: remove this once gin finishes the past meridian df
past_df = past_df.drop(past_df.columns[17:], axis=1)
past_df.shape
```

```python
past_future_df = pd.concat([past_df, future_df], axis=1)
```
